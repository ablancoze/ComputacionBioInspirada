Primero importamos las clases que utilizaremos:
Utilizaremos numpy para el manejo de arrays.
De Keras importamos el tipo de modelo Sequential 
El tipo de capa Dense que es la “normal”.

Como se puede ver son las cuatro entradas posibles de la función XOR 
[0,0], [0,1], [1,0],[1,1]
y sus cuatro salidas:
0, 1,1,0.

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense

# cargamos las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# y estos son los resultados que se obtienen, en el mismo orden
target_data = np.array([[0],[1],[1],[0]], "float32")
print(training_data)
print(target_data)
```

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
 
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
 
```

```python
model.fit(training_data, target_data, epochs=500, verbose=2)

# evaluamos el modelo
scores = model.evaluate(training_data, target_data)
print("end evaluate")
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data).round())
```

Guardar la red con sus pesos y recuperarla


```python
# serializar el modelo a JSON
model_json = model.to_json()
with open("models/nn_xor.json", "w") as json_file:
    json_file.write(model_json)
# serializar los pesos a HDF5
model.save_weights("models/nn_xor.h5")
print("Modelo Guardado!")
 
# mas tarde...
 
# cargar json y crear el modelo
json_file = open('models/nn_xor.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
```
