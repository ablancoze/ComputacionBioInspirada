```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.model_selection import train_test_split

FILENAME = './data/pima-indians-diabetes.csv'
data = pd.read_csv(FILENAME, header=None, names=['a','b','c','d','e','f','g','h','S'])

D=pd.get_dummies(data,columns=["S"])
normD=(D-D.min())/(D.max()-D.min())
normD.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(normD, test_size=0.2)
Xtrain=train.loc[:,'a':'h']
Ytrain=train.loc[:,'S_0':'S_1']
Xtest=test.loc[:,'a':'h']
Ytest=test.loc[:,'S_0':'S_1']

Xtrain
```

```python

model = Sequential()
model.add(Dense(40, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(20, kernel_initializer='uniform', activation='relu'))
model.add(Dense(5, kernel_initializer='uniform', activation='relu'))
model.add(Dense(2, kernel_initializer='uniform', activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
historic=model.fit(Xtrain, Ytrain, epochs=30, batch_size=5)
scores = model.evaluate(Xtest, Ytest)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
print(historic)
```
