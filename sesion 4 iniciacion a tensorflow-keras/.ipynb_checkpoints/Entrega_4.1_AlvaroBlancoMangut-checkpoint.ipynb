```python
import keras
import numpy as np
import pandas
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
```

```python
##carga datos
dataframe = pandas.read_csv("breast_cancer_dataset_new.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:8].astype(float)
Y = dataset[:,9]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y=np_utils.to_categorical(encoded_Y)
#print(X)
#print(Y)

##Division de datos para realizar test (control de overFitin)
#print(X.shape)
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42)
#print(X_train.shape)
#print (X_test.shape)

##En X_train almacenamos los datos para entrenar la red y en X_test almacenamos un 33% del total de los datos
##para comprobar si la red esta bien construida.
```

```python
##contruccion de la red con Keras
lr=0.04
neural_net=[8,4,8,8,4,2]
model = keras.Sequential()

#Capa 1
model.add(keras.layers.Dense(neural_net[1], activation='relu'))

#Capa 2
model.add(keras.layers.Dense(neural_net[2], activation='relu'))

#Capa 3
model.add(keras.layers.Dense(neural_net[3], activation='relu'))

#Capa 4
model.add(keras.layers.Dense(neural_net[4], activation='relu'))

#Capa 5
model.add(keras.layers.Dense(neural_net[5], activation='sigmoid'))

#Compilacion del modelo
model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=lr), metrics=['acc'])

#Entrnamos a la red con X_train e Y_train
history=model.fit(X_train, Y_train, epochs=400, validation_data=(X_test, Y_test))
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
##carga datos
dataframe = pandas.read_csv("iris.data.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y=np_utils.to_categorical(encoded_Y)
#print(X)
#print(Y)

##Division de datos para realizar test (control de overFitin)
#print(X.shape)
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42)
#print(X_train.shape)
#print (X_test.shape)

##En X_train almacenamos los datos para entrenar la red y en X_test almacenamos un 33% del total de los datos
##para comprobar si la red esta bien construida.

##contruccion de la red con Keras
lr=0.02
neural_net=[4,4,8,8,4,3]
model = keras.Sequential()

#Capa 1
model.add(keras.layers.Dense(neural_net[1], activation='relu'))

#Capa 2
model.add(keras.layers.Dense(neural_net[2], activation='relu'))

#Capa 3
model.add(keras.layers.Dense(neural_net[3], activation='relu'))

#Capa 4
model.add(keras.layers.Dense(neural_net[4], activation='relu'))

#Capa 5
model.add(keras.layers.Dense(neural_net[5], activation='softmax'))

#Compilacion del modelo
model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=lr), metrics=['acc'])
```

```python
#Entrnamos a la red con X_train e Y_train
history=model.fit(X_train, Y_train, epochs=600, validation_data=(X_test, Y_test))
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
