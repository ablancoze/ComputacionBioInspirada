```python
import keras
import numpy as np
import pandas
import math 
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
```

```python
class neural_net_pro:
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y
        
    def layer_number(self):
        return int (math.sqrt(self.X.shape[1]*self.Y.shape[1]))
```

```python
##carga datos
dataframe = pandas.read_csv("breast_cancer_dataset_new.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:8].astype(float)
Y = dataset[:,9]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y=np_utils.to_categorical(encoded_Y)
#print(X)
#print(Y)

##Division de datos para realizar test (control de overFitin)
#print(X.shape)
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42)
#print(X_train.shape)
#print (X_test.shape)

##En X_train almacenamos los datos para entrenar la red y en X_test almacenamos un 33% del total de los datos
##para comprobar si la red esta bien construida.
```

```python
##carga datos
dataframe = pandas.read_csv("sonar.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:59].astype(float)
Y = dataset[:,60]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y=np_utils.to_categorical(encoded_Y)
#print(X)
#print(Y)

##Division de datos para realizar test (control de overFitin)
#print(X.shape)
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42)
#print(X_train.shape)
#print (X_test.shape)
```

```python
##contruccion de la red con Keras
lr=0.04
neural_net=[8,4,2]
model = keras.Sequential()

#Capa 1
model.add(keras.layers.Dense(neural_net[2],input_dim=59, activation='relu'))

model.add(keras.layers.Dense(30, activation='relu'))

model.add(keras.layers.Dense(25, activation='relu'))
#Capa 3
model.add(keras.layers.Dense(neural_net[2], activation='sigmoid'))

#Compilacion del modelo
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.SGD(lr=lr), metrics=['acc'])

#Entrnamos a la red con X_train e Y_train
history=model.fit(X_train, Y_train, epochs=600, validation_data=(X_test, Y_test))
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
##carga datos
dataframe = pandas.read_csv("iris.data.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y=np_utils.to_categorical(encoded_Y)
#print(X)
#print(Y)

##Division de datos para realizar test (control de overFitin)
#print(X.shape)
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42)
#print(X_train.shape)
#print (X_test.shape)
```

```python
##En X_train almacenamos los datos para entrenar la red y en X_test almacenamos un 33% del total de los datos
##para comprobar si la red esta bien construida.

##contruccion de la red con Keras
lr=0.01

#numero_capas = math.sqrt(X.shape[1]*Y.shape[1])
#print(int(numero_capas))

neural_net=[4,5,3]
model = keras.Sequential()

#Capa 1
model.add(keras.layers.Dense(5,input_dim=4, activation='relu'))
#Capa 2
model.add(keras.layers.Dense(6, activation='relu'))
#Capa 3
model.add(keras.layers.Dense(3, activation='softmax'))

model.summary()

#Compilacion del modelo
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.SGD(lr=lr,momentum=0.01), metrics=['acc'])
```

```python
#Entrnamos a la red con X_train e Y_train
history=model.fit(X_train, Y_train,batch_size = 2,epochs=500, validation_data=(X_test, Y_test))
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
